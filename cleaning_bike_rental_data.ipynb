Combine files into single object

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

citibike_files = glob.glob('data/JC*.csv')

df_list = []

for filename in citibike_files:
    raw_data = pd.read_csv(filename)
    df_list.append(raw_data)

citibike_df = pd.concat(df_list)

citibike_df.head()
```

Convert Gender to human readable format

```python
citibike_df.describe()
citibike_df.info()
citibike_df.isna().sum()
```

```python
citibike_df["Gender"] = citibike_df["Gender"].astype(str).str.replace('1','Male').str.replace('2','Female')
citibike_df["Start Time"] = pd.to_datetime(citibike_df["Start Time"])
citibike_df["Stop Time"] = pd.to_datetime(citibike_df["Stop Time"])

# citibike_df.dtypes()
citibike_df.head()
```
